```python
# Monthly order trend
# Prepare Data
# convert column to datetime
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])
# Create Months Name
orders["Month"] = orders["order_purchase_timestamp"].dt.month_name()
# Set Correct Month Order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
orders["Month"] = pd.Categorical(
    orders["Month"],
    categories=month_order,
    ordered=True
)
data = orders.groupby("Month")["order_id"].count()
# draw line plot
data.plot(kind = "line",figsize = (6,4))
# show all month names
plt.xticks(ticks=range(len(data.index)), labels=data.index, rotation=45)
plt.grid(linestyle = "--")
plt.xlabel("Month Name")
plt.ylabel("Total Orders")
plt.title("Total Orders By Month")
plt.tight_layout()
plt.show()
```

##**RFM Analysis and Cusotmer Segmentation**

```python
# Performing RFM
# Merge tables of perform RFM
df = orders.merge(order_payments, on = "order_id")
df = df.merge(customers, on= "customer_id")

# finding last order date in dataset
last_active_date = df["order_purchase_timestamp"].max()

#Finding RFM value using lambda function
rfm = df.groupby("customer_id").agg({"order_purchase_timestamp": lambda x: (last_active_date - x.max()).days,
                                         "order_id": "count",
                                         "payment_value": "sum"})
# rename columns
rfm.columns = ["recency","frequency","monetery"]
rfm.head()
```

```python
# Score Customers
rfm['r_score'] = pd.qcut(rfm['recency'], 4, labels=[4,3,2,1])
rfm['f_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 4, labels=[4,3,2,1])
rfm['m_score'] = pd.qcut(rfm['monetery'], 4, labels=[1,2,3,4])

# create segments
rfm["segment"] = rfm["r_score"].astype(str)+ rfm["f_score"].astype(str)


# make segment more clear
def segment_customer(row):
    if row['r_score'] >= 3 and row['f_score'] >= 3:
        return 'High Value'
    elif row['r_score'] >= 2 and row['f_score'] >= 2:
        return 'Regular'
    else:
        return 'At Risk'

rfm['customer_Type'] = rfm.apply(segment_customer, axis=1)
rfm.head()
```

```python
# Pie chart for customer segementation
rfm["customer_Type"].value_counts().plot(kind = "pie",autopct='%1.1f%%')
plt.title("Customer_type")
plt.ylabel("")
plt.show()
```
