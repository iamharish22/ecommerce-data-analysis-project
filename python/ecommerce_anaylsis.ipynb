```python
# checking any column contains null values
print(df_orders.isnull().sum())
# orders dataset contains some null values as they are time series data we can fill them using forward or backword fill
df_orders = df_orders.ffill()
# succesfully handalled missing values
print(df_orders.isnull().sum())
```

```python
# merge tables
# four tables are orders,customers,payments and order items so we need to merge order and customers and payment and orders 
df_cus_or = df_customer.merge(df_orders, on = "customer_id") 
df_cus_or.head(2)
```

```python
#merge payment and orders tables
df_or_pay = df_orders.merge(df_orderspayments, on = "order_id")
df_or_pay.head(2)
```

```python
# Monthly order trend
# convert column to datetime
df_orders["order_purchase_timestamp"] = pd.to_datetime(df_orders["order_purchase_timestamp"])
data = df_orders.groupby(df_orders["order_purchase_timestamp"].dt.month_name())["order_id"].count()
# draw line plot
data.plot(kind = "line",figsize = (6,4))
# show all month names
plt.xticks(ticks=range(len(data.index)), labels=data.index, rotation=45)
plt.grid(linestyle = "--")
plt.xlabel("Month Name")
plt.ylabel("Total Orders")
plt.title("Total Orders By Month")
plt.tight_layout()
plt.show()
```
